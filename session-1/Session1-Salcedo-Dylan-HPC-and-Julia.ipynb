# KR1: Julia @code_* macros

The `code_*` macros allow us to inspect what Julia does "under the hood" by showing the assembly operations needed to carry out the Julia operations. For our example, we shall take the power operation (raising a number 3 to the power 2). We will be using macros (commands that start with `@`) to directly infer to the operation following it.

- `code_lowered` shows us a the lower level Julia operations that constitute the operation. In our case, Julia will use its `Base.power_by_squaring` operation to carry out the code. 

```julia
@code_lowered 3^2
```

- `code_typed`, is a similar macro to `code_lowered`, but since Julia invokes type inference before compiling (see KR3 for type inferencing), this macro shows us the data types being used for each operation 

```julia
@code_typed 3^2
```

- `code_llvm` shows us the assembly operations in LLVM-IR, or the type of compiler specific to our implementation of Julia.

```julia
@code_llvm 3^2
```

- `code_native` shows us the assembly operations that are specific to our type of device. This one represents the actual commands given to the machine (in assembly form, so it doesn't show us 1s and 0s). In this case, I am using a x84_64 CPU.

```julia
@code_native 3^2
```

- `code_warntype` shows us the value of Julia's type inference. In our example, the value of y (i.e. what the pos() function returns) can be of any data type. Hoever, the function f(x) is strict in terms of the type of its output, which may raise an error if pos() would return a Int64 type.  

```julia
pos(x) = x < 0 ? 0 : x

function f(x)
   y = pos(x)
   return sin(y*x + 1)
end

@code_warntype f(3.2)
```

# KR2: Constants in Julia
Julia can directly determine constants produced by its built-in functions. In the example below, we know that the output of the sine function is a float. Instead of Julia creating another operation to evaluate the sine function, it tells the compiler that this output is already a float.

```julia
@code_typed 3.5*6
```

```julia
@code_typed sin(4)*3
```

# KR3: Demonstrate Julia's type-inference and multiple dispatch
Julia has type inference when it operates on variables. It compiles several versions of the code (one for each variable type), then adds the appropriate compilation onto the final execution. This way, Julia won't need to check for the variable type for each operation.

```julia
@code_native 3^2
```

```julia
@code_native 3.5^2
```

Each operation in Julia can have several forms, which we call methods. For example, Julia's `sin()` function has a separate method for when its argument is a float and a separate method for when its argument is a complex number. To see all the methods Julia has for each operation, we can call the `method()` function. 

```julia
methods(sin)
```

```julia
@which sin(1)
```

```julia
@which sin(1.0)
```

In this aspect, the main problem that a programming language must solve is to find the proper method to use - this is called dispatching. Julia excels in dispatching by employing a particular process called "multiple dispatching". Using type-inferencing, Julia is able to dispatch the proper method with very minimal effect on processing power. 

# KR4: Time difference of built-in functions and created functions
We shall use the `@time` macro to show the time it takes to process our custom function that takes the sum of all elements in the vector.

```julia
function mysum(x::Vector)
    s = zero(eltype(x))
    for el in x
        s += el
    end
    return s
end


q = rand(10_000_000)
@time mysum(q)
```

Running the same operation would then result in less allocation in memory because the operation has already been compiled

```julia
@time mysum(q)
```

```julia
r = rand(10_000_000)
@time mysum(r)
```

Doing the same operation for Julia's built-in `sum()` function, we see that execution time is less than that of our custom function, disregarding compilation time.

```julia
@time sum(q)
```

```julia
@time sum(q)
```

```julia
@time plotMandelbrot( (-1.0,1.0), n=10_000 );
```

# KR6: Computational complexity of the Mandelbrot plot
We measure the time it takes for Julia to complete the Mandelbrot plot simulation as a function of n, or the number of iterations to compute each point in the Mandelbrot function

```julia
times = []
for i in range(100,10000,length=100)
    itime = @timed plotMandelbrot( (-1.0,1.0), n=Int(i) )
    push!(times,(i,itime.time))
end
```

```julia
plot(
    getindex.(times,1),
    getindex.(times,2),
    lw = 3,
    title = "Computational complexity of the Mandelbrot plotter",
    label="",
    xlabel="n",
    ylabel="Time (s)"
)
```

# KR7: Discussion on computational complexity
As we can see, the Mandelbrot plot function takes on a computational complexity of `O(log(n))`, meaning that for every 10^k times we increase n, the time it takes to complete the plot increases by k-times. 
We know that the Mandelbrot plot forms a fractal, where the features of smaller components are similar to that of the larger components. Therefore, the "resolution" of the fractal is determined by the value of n, where greater values of n provide a better approximation of each point's value.
